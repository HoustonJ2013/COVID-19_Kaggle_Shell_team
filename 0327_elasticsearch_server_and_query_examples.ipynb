```python
from datetime import datetime
from elasticsearch import Elasticsearch, helpers
es = Elasticsearch(
    hosts=[{'host': "99.44.95.175", 'port': 9200}])
```

```python
## Query Examples
```

```python
## Query to get all the ids
query = { 
    "query" : { 
        "match_all" : {} 
    },
    "stored_fields": []
}
a=helpers.scan(es,query=query,scroll='1m',index='covid-kaggle')#like others so far
all_es_ids=[aa['_id'] for aa in a]
print("There are %i of jsons"%(len(all_es_ids)))
```

```python
## Query example: Search Authors with last name Liu
doc = {
        'size' : 50,
        'query': {
           "match": {"metadata.authors.last": "Liu", 
                    }
        }
   }
res1 = es.search(index='covid-kaggle', doc_type='pdf_json', body=doc)
res1["hits"]["total"] , res1["hits"]["hits"][0]["_source"]["metadata"]["authors"][0]
```

```python
## Check one record 
record_example = res1["hits"]["hits"][0]
record_example
```

```python
## Query Example: Search title with risk
doc = {
        'size' : 50,
        'query': {
           "match": {"metadata.title": "risk"}
        }
   }
res2 = es.search(index='covid-kaggle', doc_type='pdf_json', body=doc)
res2["hits"]["total"] , res2["hits"]["hits"][0]["_source"]["metadata"]["title"]
```

```python
## Query Example: Search body text with "corona", "virus", "risk"
doc = {
        'query': {
            "bool":{
                "must":
                   [{"match": {"body_text.text": "corona"}},  
                    {"match": {"body_text.text":"virus"}}, 
                    {"match": {"body_text.text":"risk"}}
                    ], 
            }
        }
   }
res3 = es.search(index='covid-kaggle', doc_type='pdf_json', body=doc, size=50)
res3["hits"]["total"] , res3["hits"]["hits"][0]["_source"]["abstract"]
```
